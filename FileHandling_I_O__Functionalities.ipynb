```python
import csv
from datetime import datetime  # For timestamp in report.

filename = 'data' # Update to your CSV path, e.g., 'SampleSuperstore.csv'.

with open(filename, 'r', newline='') as file:
    reader = csv.reader(file)
    headers = next(reader)  # Column names.

    rows = list(reader)
    total_rows = len(rows)

    # Count empties per column.
    empty_counts = {header: 0 for header in headers}
    for row in rows:
        for i, value in enumerate(row):
            if value == '':  # Manual empty check.
                empty_counts[headers[i]] += 1

    # Summary stats.
    total_empties = sum(empty_counts.values())
    avg_empties = total_empties / len(headers) if headers else 0

# Write human-readable report.
report_filename = 'csv_report.txt'
with open(report_filename, 'w') as report:
    report.write(f"CSV Analysis Report\n")
    report.write(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    report.write(f"File: {filename}\n")
    report.write(f"{'='*50}\n\n")

    report.write(f"Dataset Summary:\n")
    report.write(f"- Total data rows: {total_rows:,}\n")
    report.write(f"- Columns: {len(headers)}\n")
    report.write(f"- Total empty cells: {total_empties:,}\n")
    report.write(f"- Avg empties per column: {avg_empties:.1f}\n\n")

    report.write("Empty Values per Column:\n")
    report.write(f"{'Column':<25} {'Empties':<10} {'%':<8}\n")
    report.write("-" * 45 + "\n")
    for header, count in empty_counts.items():
        pct = (count / total_rows * 100) if total_rows > 0 else 0
        report.write(f"{header:<25} {count:<10,} {pct:<7.1f}%\n")

print(f"Report saved: {report_filename}")
print("Sample output preview:")
print(f"Total rows: {total_rows:,}")
print(f"Total empties: {total_empties:,}")
```
